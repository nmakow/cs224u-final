```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
import io
from utils import build_rnn_dataset
```

```python
plt.rcParams["figure.figsize"] = (10,5)
data_path = "./data/"
```

```python
train = pd.read_csv(data_path + "train.csv")
num_train = train.shape[0]
print("%d train examples." % num_train)
```

```python
X_rnn, Y_rnn = build_rnn_dataset(train, 1)
```

```python
print(len(X_rnn['train']))
```

```python
char_counts = defaultdict(int)
counter = 1
for comment in X_rnn['train']:
    for token in comment:
        for char in token:
            char_counts[char.encode("unicode-escape")] += 1
    if counter % 10000 == 9999:
        print("Processed %d examples..." % (counter))
    counter += 1
print(len(char_counts.keys()))
```

```python
print(char_counts)
```

```python
trimmed_char_counts = defaultdict(int)
threshold = 25
for char,count in char_counts.items():
    if count >= threshold:
        trimmed_char_counts[char] = count
print(len(trimmed_char_counts))
```

```python
tmp = sorted(list(trimmed_char_counts.items()), key=lambda x: -x[1])
```

```python
sorted_chars = [x[0].decode('unicode-escape') for x in tmp]
sorted_counts = [x[1] for x in tmp]
```

```python
print([x[0] for x in tmp])
```

```python
fig=plt.figure(figsize=(50, 20), dpi=80, facecolor='w', edgecolor='k')
plt.yscale('log')
plt.bar(np.arange(len(sorted_chars)[:50]), sorted_counts[:50], tick_label=sorted_chars[:50])
```
